```python
# Crear los diccionarios individuales
diccionarios_por_estado = {}

for estado in estados:
    estado_nombre = estado['estado'].replace(" ", "_")  #quitar los espacios en las keys, tambien se podrian quitar los acentos
    #descomponemos las claves y los valores en dos listas  por cada uno de los estados
    claves = [f"{clave}_{estado_nombre}" for clave in estado.keys()]
    print(claves)
    valores = estado.values()
    print(valores)
    diccionario_estado = dict(zip(claves, valores)) #
    diccionarios_por_estado[f"diccionario_{estado_nombre}"] = diccionario_estado

# Mostramos el resultado
for nombre_diccionario, diccionario in diccionarios_por_estado.items():
    print(f"{nombre_diccionario}: {diccionario}\n")
```

antiguo y el más moderno (es necesario que sea dinámico y que el programa sirva
para 3 estados como para 50). ¿Cuántos años de diferencia hay entre ellos?

```python
import datetime
def estado_mas_antiguo_mas_moderno(diccionario_dinamico_estados): #al recibir un parametro diccionario de estados, la funcion sirve para (n) estados
    # Obtener las fechas de fundación de todos los estados
    fechas_fundacion = []
    for nombre_diccionario, diccionario in diccionario_dinamico_estados.items():
        fecha_fundacion = datetime.datetime.strptime(diccionario['Fecha_fundacion_' + nombre_diccionario.split('diccionario_')[1]], '%d-%m-%Y')
        fechas_fundacion.append(fecha_fundacion)

    # Encontrar la fecha de fundación más antigua y más moderna
    fecha_fundacion_mas_antigua = min(fechas_fundacion)  # min y max funcionan con datetime y otros tipos
    fecha_fundacion_mas_moderna = max(fechas_fundacion)

    # Obtener los nombres de los estados correspondientes a las fechas de fundación más antigua y más moderna
    for nombre_diccionario, diccionario in diccionario_dinamico_estados.items():
        fecha_fundacion = datetime.datetime.strptime(diccionario['Fecha_fundacion_' + nombre_diccionario.split('diccionario_')[1]], '%d-%m-%Y')
        if fecha_fundacion == fecha_fundacion_mas_antigua:
            estado_mas_antiguo = nombre_diccionario.split('diccionario_')[1]
        if fecha_fundacion == fecha_fundacion_mas_moderna:
            estado_mas_moderno = nombre_diccionario.split('diccionario_')[1]

    # Calcular la diferencia en años entre las fechas de fundación
    diferencia_anios = (fecha_fundacion_mas_moderna - fecha_fundacion_mas_antigua).days / 365


    # Imprimir los resultados
    print(f"El estado más antiguo es {estado_mas_antiguo} y fue fundado el {fecha_fundacion_mas_antigua.strftime('%d-%m-%Y')}")
    print(f"El estado más moderno es {estado_mas_moderno} y fue fundado el {fecha_fundacion_mas_moderna.strftime('%d-%m-%Y')}")
    print(f"La diferencia de años entre ellos es de {diferencia_anios:.0f} años")

estado_mas_antiguo_mas_moderno(diccionarios_por_estado) #al pasarle como argumento un diccionario de estados la funcion sirve para (n) estados
```

Alabama y South Carolina son dos números aleatorios entre 0 y 0.1. Asignaremos a

¿Cuántos años tardará South Carolina en alcanzar en población a Alabama? Vamos
a asumir la tasa de fallecidos y la población correspondientes a 2021.

```python
# Poblaciones de 2001
import random
poblacion_alabama_2001 = 4451493
poblacion_south_carolina_2001 = 4023438

# Generar tasas de crecimiento aleatorias
tasas_crecimiento = [random.uniform(0, 0.1) for i in range(2)]  #sorted sirve para ordenar los resultados de menor a mayor, et random.uniform permite definir los rangos definidos
# https://www.w3schools.com/python/ref_func_sorted.asp
tasas_crecimiento =sorted(tasas_crecimiento)  #probar con , reverse=True) como se van los años, con el riesgo de que el while no se cumpla
tasa_crecimiento_alabama = tasas_crecimiento[0]
tasa_crecimiento_south_carolina = tasas_crecimiento[1]

# a. ¿Cuántos años tardará el estado de South Carolina en alcanzar en población a Alabama?
# b. ¿En qué año ocurrirá esto?
anio_actual = 2001 #se inicializa el conteo desde el año en cuestión
while poblacion_south_carolina_2001 < poblacion_alabama_2001:  # iterar hasta que South Carolina alcance la población de Alabama
    poblacion_alabama_2001 += poblacion_alabama_2001 * tasa_crecimiento_alabama
    poblacion_south_carolina_2001 += poblacion_south_carolina_2001 * tasa_crecimiento_south_carolina
    anio_actual += 1 #si no alcanzó el año sumarle uno e iterar

anio_alcance = anio_actual
anios_necesarios = anio_actual - 2001 # almacenamos la diferencia de años

print(f"a. South Carolina tardará {anios_necesarios} años en alcanzar a Alabama en población.\nb. Esto ocurrirá en el año {anio_alcance}")


# Poblaciones y tasas de fallecidos de 2001 (considerando las tasas de fallecidos de 2021)
poblacion_alabama_2001 = 4451493
poblacion_south_carolina_2001 = 4023438
# Para calcular la tasa de mortalidad tenemos que dividir el número de muertes
# en un periodo entre la población media en dicho periodo y multiplicar el
# resultado por mil.
tasa_fallecidos_alabama = 15912 / (4451493/2) * 1000   # no hace falta el paréntesis, es solo por aclarar
tasa_fallecidos_south_carolina = 9500 / (4023438/2) * 1000

# c: Calcular el año en que South Carolina alcanzará a Alabama considerando las tasas de fallecidos
anio_actual = 2001
while poblacion_south_carolina_2001 < poblacion_alabama_2001:
    poblacion_alabama_2001 += poblacion_alabama_2001 * (tasa_crecimiento_alabama - tasa_fallecidos_alabama)
    poblacion_south_carolina_2001 += poblacion_south_carolina_2001 * (tasa_crecimiento_south_carolina - tasa_fallecidos_south_carolina)
    anio_actual += 1

anio_alcance_con_fallecidos = anio_actual
anios_necesarios_con_fallecidos = anio_actual - 2001

print(f"c. South Carolina tardará {anios_necesarios_con_fallecidos} años en alcanzar a Alabama en población considerando las tasas de fallecidos.")
print(f"Esto ocurrirá en el año {anio_alcance_con_fallecidos} considerando las tasas de fallecidos.")
```

# Ejercicio 3

Cread una función que genere una proyección para el año 2002, utilizando como ratio la
comparativa entre los años 2000 y 2001. De tal forma:
Población Año 2002 = Población Año 2001 / Población Año 2000 x Población año 2001.
Tras un estudio demográfico se ha determinado que el número de habitantes de cierta
población, en los próximos años, vendrá dado por la función: P(t) = 14500t + 7000/2t + 1
donde t son los años transcurridos entre la fundación del Estado y 1900.
En la misma función, devolved la población que tendría cada Estado bajo la estimación que
devuelve ese estudio demográfico.
Añadid los resultados a cada diccionario de la lista.

```python
def proyeccion_poblacion_2002(diccionarios_por_estado):

  for nombre_diccionario, diccionario in diccionarios_por_estado.items():
    # Calcular la proyección para el año 2002
    poblacion_2000 = diccionario.get('poblacion_2000_' + nombre_diccionario.split('diccionario_')[1], 0)
    poblacion_2001 = diccionario.get('poblacion_2001_' + nombre_diccionario.split('diccionario_')[1], 0)
    poblacion_proyectada_2002 =  ( poblacion_2001 / poblacion_2000 * poblacion_2001)  # no está claro el enunciado pero la proyeccion es igual al ultimo año * la tasa de crecimiento mencionada en el ejercicio
    diccionario['poblacion_proyectada_2002_' + nombre_diccionario.split('diccionario_')[1]] = int(poblacion_proyectada_2002)

    # Obtener la fecha de fundación del estado
    fecha_fundacion = datetime.datetime.strptime(diccionario['Fecha_fundacion_' + nombre_diccionario.split('diccionario_')[1]], '%d-%m-%Y')

    # Calcular la diferencia entre la fecha de fundación y 1900
    diferencia_anios = 1900 - fecha_fundacion.year
    print(diferencia_anios)

    # Calcular la población estimada por la función
    poblacion_estimada = int(14500 * diferencia_anios + (7000 / (2 * diferencia_anios))+ 1)
    print(poblacion_estimada)
    diccionario['poblacion_estimada_' + nombre_diccionario.split('diccionario_')[1]] = poblacion_estimada

  return diccionarios_por_estado

diccionarios_por_estado = proyeccion_poblacion_2002(diccionarios_por_estado)

for nombre_diccionario, diccionario in diccionarios_por_estado.items():
  print(f"{nombre_diccionario}: {diccionario}\n")
```

Inicializa el mapa en las coordenadas:
Latitud: 30.101271

Para el tamaño de los círculos, toma como radio el resultado de dividir la población de 2002
entre 100:
Radio = Población en 2002 / 100
Nota: Es posible que Jupyter Notebook entregue un mensaje diciendo que debéis verificar
el notebook para poder incluir el mapa. Pulsad en File -> Trusted Notebook para verificarlo

```python
import folium
#evaluar si hay que usar esta funcion ya que los campos de coordenadas parecen estar en coordenadas DMS, grados minutos segundos separados por un punto y deberian estar calculados en Grados Decimales.
def dms_a_decimal(dms):
    partes = dms.split('.')
    grados = int(partes[0])
    minutos = int(partes[1])
    segundos = float(partes[2])

    decimal = grados + (minutos / 60) + (segundos / 3600)
    return decimal

# Ejemplo de uso
latitud_dms = '33.258.882'
longitud_dms = '-86.829.534'

latitud_decimal = dms_a_decimal(latitud_dms)
longitud_decimal = dms_a_decimal(longitud_dms)

print(f"Latitud decimal: {latitud_decimal}")
print(f"Longitud decimal: {longitud_decimal}")


# Proyección de la población para 2002
for estado in estados:
    # Calcular la tasa de crecimiento de 2000 a 2001
    tasa_crecimiento = (estado['poblacion_2001'] - estado['poblacion_2000']) / estado['poblacion_2000']

    # Proyectar la población para 2002
    poblacion_2002 = estado['poblacion_2001'] * (1 + tasa_crecimiento)

    # Agregar la población proyectada a los datos del estado
    estado['poblacion_2002'] = poblacion_2002
    print(estado)
# Inicializar el mapa en las coordenadas especificadas
mapa = folium.Map(location=[30.101271, -82.370146], zoom_start=6)

# Añadir marcadores al mapa usando bucles
for estado in estados:
    # Coordenadas del estado
    lat = float(estado['latitud'])
    lon = float(estado['longitud'])

    # Población proyectada y radio del círculo
    poblacion_2002 = estado['poblacion_2002']
    radio = poblacion_2002 / 1000000

    # Añadir marcador
    folium.CircleMarker(
        location=(lat, lon),
        radius=radio,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{estado['estado']}: {poblacion_2002:.0f} habitantes en 2002"
    ).add_to(mapa)

mapa
```
